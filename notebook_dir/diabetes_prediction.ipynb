**Data Preprocessing**

```python
#separating data into features and targets
X = df.drop(columns = [TARGET_COL])
y = df[TARGET_COL]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.2,
    stratify = y,
    random_state = RANDOM_STATE
)

print("Dataset size", X.shape)
print("Train Dataset size", X_train.shape)
print("Test Dataset size", X_test.shape)
```

**preprocessing pipeline**

```python
#0->NaN for selected columns: Nan -> median imputation
zero_as_missing_cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

numerical_features = X_train.select_dtypes(include = [np.number]).columns.tolist()

def replace_zeros_with_nan_df(X, cols):
    X = X.copy()
    for col in cols:
        if col in X.columns:
            X[col] = X[col].replace(0, np.nan)
    return X

zero_to_nan = FunctionTransformer(
    replace_zeros_with_nan_df,
    kw_args = {"cols": zero_as_missing_cols},
    validate = False

)

numerical_transformer = Pipeline(steps=[
    ("zero_to_nan", zero_to_nan),
    ("imputer", SimpleImputer(strategy = "median")),
    ("scaler", StandardScaler())
])

preprocess = ColumnTransformer(
    transformers = [
        ("num", numerical_transformer ,numerical_features)
    ]
)
```

**Baseline Model - Logistic Regression**

```python
baseline_lr = Pipeline(
    steps = [
        ("preprocess", preprocess),
        ("model", LogisticRegression())
    ]
)
```

```python
baseline_lr.fit(X_train, y_train)
```

```python
#reusable evaluation function
def evaluate_classifier(model, X_train, y_train, X_test, y_test, model_name):
    #predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    #accuracy
    train_acc = accuracy_score(y_train, y_train_pred)*100
    test_acc = accuracy_score(y_test, y_test_pred)*100

    #results
    print(f"print {model_name} - Train Accuracy: {train_acc:.2f}%")
    print(f"print {model_name} - Test Accuracy: {test_acc:.2f}%\n")

    print("-"*40)

    print("Train Classification Report")
    print(classification_report(y_train, y_train_pred))

    print("-"*40)

    print("Test Classification Report")
    print(classification_report(y_test, y_test_pred))
```

```python
evaluate_classifier(
    baseline_lr,
    X_train,
    y_train,
    X_test,
    y_test,
    model_name = "Baseline Logistic Regression"
)
```

**Model Selection and Optimization**

```python
#scale_pos_weight : negative samples/positive samples

neg =(y_train==0).sum()
pos =(y_train==1).sum()
scale_pos_weight = neg/max(pos,1)

print("scale_pos_weight", scale_pos_weight)
```

```python
#models to try
models ={
    "LogisticRegression": Pipeline(
        [
            ("preprocess", preprocess),
            ("model", LogisticRegression(
                random_state=RANDOM_STATE,
                class_weight = "balanced",
                max_iter=1000
            ))
        ]
    ),
     "SVC": Pipeline(
        [
            ("preprocess", preprocess),
            ("model", SVC(
                class_weight="balanced",
                random_state=RANDOM_STATE
            ))
        ]
    ),
    "RandomForest": Pipeline(
        [
            ("preprocess",preprocess),
            ("model", RandomForestClassifier(
                random_state=RANDOM_STATE,
                class_weight="balanced_subsample"

            ))
        ]
    ),
    "XGBClassifier": Pipeline(
        [
            ("preprocess", preprocess),
            ("model", XGBClassifier(
                scale_pos_weight=scale_pos_weight,
                eval_metric="logloss",

            ))
        ]
    )
}
```

```python
from sklearn.model_selection import StratifiedKFold
k = 5
cv = StratifiedKFold(n_splits=k, shuffle=True, random_state = RANDOM_STATE)
```

```python
for name, model_pipeline in models.items():
    recall_scores = []
    f1_scores = []

    for tr_idx, te_idx in cv.split(X_train, y_train):
        X_tr, X_te = X_train.iloc[tr_idx], X_train.iloc[te_idx]
        y_tr, y_te = y_train.iloc[tr_idx], y_train.iloc[te_idx]

        model_pipeline.fit(X_tr, y_tr)
        pred=model_pipeline.predict(X_te)

        recall_scores.append(round(recall_score(y_te, pred), 2))
        f1_scores.append(round(f1_score(y_te, pred), 2))

    print("Model Name:", name)
    print("Recall Scores:", recall_scores)
    print("f1 scores:", f1_scores)
    print("\nRecall Mean:", round(float(np.mean(recall_scores)),4))
    print("f1 score Mean:", round(float(np.mean(f1_scores)),4))
    print("-"*50)
```

**SVC provides the highest performance**

--hyperparameter tuning

```python
svc_opt = Pipeline([
    ("preprocess", preprocess),
    ("model", SVC(
        class_weight= "balanced",
        random_state=RANDOM_STATE
    ))
])
```

```python
#parameter combinations
param_grid_svc = {
    "model__C" : [0.1,1,10,100,1000],
    "model__gamma" : [0.001, 0.01, 0.1, 1, "scale"]
}
```

```python
#grid search
grid_search = GridSearchCV(
    estimator = svc_opt,
    param_grid = param_grid_svc,
    scoring = "f1",
    cv=cv,
    verbose = True,
    n_jobs = -1
)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Best Hyperparameter Tuning Results")
print("best parameters:", grid_search.best_params_)
print("best score:", grid_search.best_score_)
```

```python
#retraining with best parameters
svc_best = Pipeline(
    [
        ("preprocess", preprocess),
        ("model", SVC(
            random_state=RANDOM_STATE,
            class_weight="balanced",
            C=1,
            gamma="scale"
        ))
    ]
)
```

```python
#training best pipeline
svc_best.fit(X_train, y_train)
```

**Model Evaluation**

```python
evaluate_classifier(
    model = svc_best,
    X_train = X_train,
    y_train = y_train,
    X_test = X_test,
    y_test = y_test,
    model_name = "Best SVC"
)
```

**Building a Prediction Model**

```python
def predict_diabetes(input_features):
    input_df = pd.DataFrame(
        [input_features],
        columns = X_train.columns
    )
    prediction = svc_best.predict(input_df)
    print("Model Prediction:", prediction)
    if prediction[0] == 1:
        print("Diagnosis - Diabetic🔴")
    else:
        print("Diagnosis - Non-Diabetic🟢")
```
